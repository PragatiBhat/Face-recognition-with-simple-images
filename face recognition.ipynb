```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
frame2=cv2.imread("joey.jpg")
```

```python
rgb2=cv2.cvtColor(frame2,cv2.COLOR_BGR2RGB)
```

```python
classifier=cv2.CascadeClassifier("haarcascade_frontalface_default.xml")
```

```python
faces1=classifier.detectMultiScale(rgb2)
```

```python
face1=faces1[0]
```

```python
x,y,w,h=face1
```

```python
out1=cv2.rectangle(rgb2,(x,y),(x+w,y+h),(0,0,255),4)
```

```python
plt.imshow(out1)
```
